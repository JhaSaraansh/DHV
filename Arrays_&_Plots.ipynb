```python
x=[1,2,3,4]
y=[4,5,6,7]
fonts=[100,200,300,700]
plt.scatter(x,y,color='r',marker='*',edgecolor='g',s=fonts)
plt.show()
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
fonts=[100,200,300,700]
colors=['r','g','b','y']
plt.scatter(x,y,color=colors,marker='*',s=fonts)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=['r','y','b','g','b']
sizes=[100,200,300,400,150]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker='*',edgecolor='g')
plt.title("Scatter",fontsize=20)
plt.xlabel('x axis',fontsize=20)
plt.ylabel('y axis',fontsize=20)
plt.show()
```

```python
x=[1,2,2,3,4,4,5,5]
plt.hist(x, color='b', edgecolor='r', bottom=1, histtype='bar', orientation='vertical')
plt.show()
```

```python
x=[10,20,30,40,50]
l=['apples','bananas','cherries','oranges','mangoes']
Ex=[0.0, 0.0, 0.0, 0.0, 0.2]
plt.pie(x,labels=l, explode=Ex)
plt.show()
```

```python
section=['a', 'b', 'c']
student=[10,20,30]
marks=[10, 20, 20, 30, 40, 40]
plt.subplot(2,2,1)
plt.bar(section, student)
plt.subplot(2,2,2)
plt.pie(student)
plt.subplot(2,2,3)
plt.hist(marks)
plt.subplot(2,2,4)
plt.plot([1,2,3],[4,5,6])
plt.show()
```
