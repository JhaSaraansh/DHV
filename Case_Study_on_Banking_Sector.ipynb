```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    'Age-Group':['18-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60', '61+'],
    'Total No. of Loans':[10, 18, 27, 21, 15, 34, 32, 12, 8, 2],
    'No. of good Loans':[3, 13, 21, 18, 13, 33, 29, 12, 7, 2],
    'No. of bad Loans':[7, 5, 6, 3, 2, 1, 3, 0, 1, 0]
}
df=pd.DataFrame(data)
print(df)
```

```python
df['Percentage of Good Loans']=(df['No. of good Loans']/df['Total No. of Loans'])*100
df['Percentage of Bad Loans']=(df['No. of bad Loans']/df['Total No. of Loans'])*100
print(df)
```

```python
df.describe()
```

```python
plt.scatter(df['No. of good Loans'], df['No. of bad Loans'])
plt.xlabel('Good Loans')
plt.ylabel('Bad Loans')
plt.title('Scatter Plot of Good Loans vs. Bad Loans')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Age-Group'], df['Total No. of Loans'], marker='o', linestyle='-', color='b', label='Value')
plt.title('Comparison of Age group vs Total no of Loans')
plt.xlabel('Age-Group')
plt.ylabel('Total Loans')
plt.grid(True)
plt.legend()
plt.show()
```

```python
correlation = df['Total No. of Loans'].corr(df['No. of good Loans'])
print("Pearson Correlation Coefficient:", correlation)
```
