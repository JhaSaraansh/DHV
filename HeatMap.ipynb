```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Heatmaps
data=np.random.random((13,16))
plt.imshow(data)
plt.title("2-D Heat Map")
plt.colorbar()
plt.show()
```
