```python
mymap=folium.Map(location=[12.9716, 77.5946],
                 width=950,
                 height=550,
                 zoom_start=12,
                 tiles='openstreetmap')
tooltip_1 = "This is Cubbon Park"
tooltip_2 ="This is Bangalore Palace"
tooltip_3 ="This is Kempegowda International Airport"
tooltip_4 ="This is Nandi Hills"

folium.Marker(
    [12.9779, 77.5952], popup="Cubbon Park", tooltip=tooltip_1).add_to(mymap)

folium.Marker(
    [13.0035, 77.5891], popup="Bangalore Palace", tooltip=tooltip_2).add_to(mymap)

folium.Marker(
    [13.1989, 77.7069], popup="Airport", tooltip=tooltip_3).add_to(mymap)

folium.Marker(
    [13.3702, 77.6835], popup="Nandi Hills", tooltip=tooltip_4).add_to(mymap)
mymap
```
