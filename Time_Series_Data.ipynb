```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

**TIME SERIES WITH TREND**

```python
data={
    "Date":pd.to_datetime(["2022-01-01", "2022-02-01", "2022-03-01", "2022-04-01", "2022-05-01", "2022-06-01", "2022-07-01", "2022-08-01", "2022-09-01", "2022-10-01"]),
    "Value":[10, 15, 22,28, 35, 42, 48, 55, 60, 68]
}
df=pd.DataFrame(data)
df=df.set_index("Date",inplace=False)
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Value"], marker='v', linestyle='-.')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

**TIME SERIES WITH SEASONAL DATA**

```python
data={
    "Date":pd.to_datetime(["2022-01-01", "2022-02-01", "2022-03-01", "2022-04-01", "2022-05-01", "2022-06-01", "2022-07-01", "2022-08-01", "2022-09-01", "2022-10-01"]),
    "Value":[10, 15, 25,38, 28, 32, 50, 38, 50, 55]
}
df=pd.DataFrame(data)
df=df.set_index("Date",inplace=False)
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Value"])
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

**SUBPLOTS**

```python
#plot 1:
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([3, 8, 1, 10, 12, 14, 20, 11])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SEASONAL")

#plot 2:
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([10, 20, 30, 40, 50, 60, 70, 80])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("TREND")

plt.show()

#plot 3:
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([10, 20, 10, 30, 20, 40, 30, 50])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("TREND AND SEASONAL")

plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
data = {
    "Date": pd.date_range(start='2022-01-01', periods=365),
    "Value": [20 + 10 * ((i % 30) // 7) for i in range(365)]  # Example seasonal pattern
}
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
result = seasonal_decompose(df["Value"], model='additive', period=30)
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(df.index, result.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(df.index, result.seasonal, label='Seasonality')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(df.index, result.resid, label='Trend and Seasonality')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
