```python
import requests

from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = "https://www.imdb.com/chart/top"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
movies = []
for row in soup.select('tbody.lister-list tr'):
  title = row.find('td', class_='titleColumn').find('a').get_text()
  year = row.find('td', class_='titleColumn').find('span', class_='secondaryInfo').get_text()[1:-1]
  rating = row.find('td', class_='ratingColumn imdbRating').find('strong').get_text()
  movies.append([title, year, rating])
time.sleep(1)
```

```python
df = pd.DataFrame(movies, columns=['Title', 'Year', 'Rating'])
df.to_csv('products.csv', index=False, encoding='utf-8')
```

```python
import requests
from bs4 import BeautifulSoup
import os

def extract_images(url, folder_path):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        img_tags = soup.find_all('img')

        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

        for img in img_tags:
            img_url = img['src']
            if not img_url.startswith('http'):
                img_url = url + '/' + img_url.lstrip('/')
            img_name = img_url.split('/')[-1]
            img_path = os.path.join(folder_path, img_name)
            img_response = requests.get(img_url)
            with open(img_path, 'wb') as img_file:
                img_file.write(img_response.content)
            print(f"Image '{img_name}' saved successfully.")
    else:
        print("Failed to retrieve the webpage.")

url = 'https://en.wikipedia.org/wiki/List_of_best-selling_automobiles'
folder_path = 'images'
extract_images(url, folder_path)
```
